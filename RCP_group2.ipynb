```python
#headers
def get_parent_model(_soup):
        lists = _soup.find_all('li', class_ ='bdc-Breadcrumb_Item')
        x = []
        for item in lists[1:]:
            x.append(item.find('span').text)
        return x
    

def get_marketing_name(_soup):
    marketing_name = dom.xpath('/html/body/main/section[1]/section/div/div/div/div[2]/div/div[1]/header/p/text()')
    return marketing_name


  
def get_weight(_soup):
    weight=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[10]/div[2]/div/p/text()')
    return weight



def get_crystal(_soup):
    crystal= dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[6]/div[2]/div/p/text()')
    return crystal 


    
def get_currency(_soup):
    currency=dom.xpath('/html/body/header/div[2]/div[1]/div/div/div/div/div/select/option[1]/text()')
    return currency


# can we do it with exceptional error
def get_price(_soup,url):
    if url =='https://us.bremont.com/products/s301-bk-br':
        price=dom.xpath('//*[@id="price-template--14636979781679__main"]/div[1]/dl/div[1]/dd/span')[0].text
        
    elif url == 'https://us.bremont.com/products/mbii-king-charles-iii-limited-edition':
        price=dom.xpath('/html/body/main/section[1]/section/div/div/div/div[2]/div/div[1]/header/div/div[1]/dl/div[1]/dd/span')[0].text
        
    else:   
        price=dom.xpath('/html/body/main/section[1]/section/div/div/div/div[2]/div/div[1]/header/div[1]/div[1]/dl/div[1]/dd/span/text()')

    return price



def get_image_URL(_soup , url):
    image_div = _soup.find('div' , class_ = 'rsp-Image rsp-Image-contain')
    image = image_div.find('img', class_ = 'rsp-Image_Image')
    url = image.get('src')   

    return url


def get_short_description(_soup):
    if url == 'https://us.bremont.com/products/s302-jet-r-s':
        short_description=dom.xpath('/html/body/main/section[1]/section/div/div/div/div[2]/div/div[4]/p[2]/text()')
    elif url=='https://us.bremont.com/products/mbii-savanna':
        short_description=dom.xpath('/html/body/main/section[1]/section/div/div/div/div[2]/div/div[3]/p/text()')
    else:
        short_description=soup.find_all(class_='prd-Product_ShortDesc fz-14_24 rte-RichText')[0].select('div p:nth-of-type(2)')[0].text
    return short_description


def get_case_material(_soup):
    case_material= dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[3]/div[2]/div/p/text()')
    return case_material


def get_caseback(_soup):
    caseback=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[4]/div[2]/div/p/text()')
    return caseback


def get_diameter(_soup):
    diameter=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[3]/div[2]/div/p/text()')
    return diameter


def get_between_lugs(_soup):
    between_lugs=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[3]/div[2]/div/p/text()')
    return between_lugs

def get_lug_to_lug(_soup):
    lug_to_lug=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[3]/div[2]/div/p/text()')
    return lug_to_lug

def get_case_thickness(_soup):
    case_thickness=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[3]/div[2]/div/p/text()')
    return case_thickness

def get_bezel_material(_soup):
    bezel_material=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[3]/div[2]/div/p/text()')
    return bezel_material

def get_bezel_color(_soup):
    bezel_color=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[3]/div[2]/div/p/text()')
    return bezel_color

def get_dial_color(_soup):
    dial_color=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[5]/div[2]/div/p/text()')
    return dial_color

def get_numerals(_soup):
    numerals=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[5]/div[2]/div/p/text()')
    return numerals


def get_bracelet_material(_soup):
    bracelet_material=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[7]/div[2]/div/p')[0].text
    return bracelet_material

def get_bracelet_color(_soup):
    bracelet_color=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[7]/div[2]/div/p')[0].text
    return bracelet_color

def get_clasp_type(_soup):
    clasp_type=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[7]/div[2]/div/p')[0].text
    return clasp_type

def get_movement(_soup):
    movement=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[1]/div[2]/div/p/text()')
    return movement

def get_caliber(_soup):
    caliber=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[1]/div[2]/div/p/text()')
    return caliber
def get_power_reserve(_soup):
    power_reserve=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[1]/div[2]/div/p/text()')
    return power_reserve
def get_frequency(_soup):
    frequency=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[1]/div[2]/div/p/text()')
    return frequency
def get_jewels(_soup):
    jewels=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[1]/div[2]/div/p/text()')
    return jewels

def get_features(_soup):
    features=dom.xpath('/html/body/main/section[4]/section/div/div/div/div/div[2]/div/div/product-accordions/ul/li[2]/div[2]/div/p/text()')
    return features

def get_description(_soup):
    description=dom.xpath('/html/body/main/section[3]/section/div/div/div/section/div/div/div/div[1]/div/div/div/p/text()')
    return description

def get_reference_number(_soup,url):
    reference_number = url.rsplit('/', 1)[-1]
    return reference_number
```

```python
# sea collection
url_ = 'https://us.bremont.com/collections/supermarine'
html = requests.get(url_)
soup_ = BeautifulSoup(html.content, "html.parser")




sea_collections = soup_.find_all('div' , class_ = 'prd-Card prd-Card-grid util-FauxLink')




    

sea_urls = []   
for item in sea_collections:
    a_tag = item.find('a', class_ = 'util-FauxLink_Link')
    href  = a_tag.get('href')
    url = "https://us.bremont.com"+href
    sea_urls.append(url)
    
print(sea_urls)
```

```python
#air collection
url_air = 'https://us.bremont.com/collections/altitude'
html = requests.get(url_air)
soup_ = BeautifulSoup(html.content, "html.parser")




air_collections = soup_.find_all('div' , class_ = 'prd-Card prd-Card-grid util-FauxLink')




    

air_urls = []   
for item in air_collections:
    a_tag = item.find('a', class_ = 'util-FauxLink_Link')
    href  = a_tag.get('href')
    url = "https://us.bremont.com"+href
    air_urls.append(url)
    
air_urls
```

```python
#land collection
url_ = 'https://us.bremont.com/collections/the-armed-forces-collection'
html = requests.get(url_)
soup_ = BeautifulSoup(html.content, "html.parser")




land_collections = soup_.find_all('div' , class_ = 'prd-Card prd-Card-grid util-FauxLink')




    
land_urls = []   
for item in land_collections:
    a_tag = item.find('a', class_ = 'util-FauxLink_Link')
    href  = a_tag.get('href')
    url = "https://us.bremont.com"+href
    land_urls.append(url)
    
land_urls
```
